```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

## pie chart

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
# pie chart top 3 countries
```

Observation : Zomata's Maximum transactions are from india then united states and Uk


```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## conclusions 
1.when rating is between 4.5-4.9 this indicates excellent                   
2.when rating is between 4.0-4.4 this indicates very good                       
3.when rating is between 3.5-3.9 this indicates good                       
4.when rating is between 3.0-3.4 this indicates average                         
5.when rating is between 2.5-2.9 this indicates average                      
6.when rating is between 2.0-2.4 this indicates poor                          


```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y='Rating count',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y='Rating count',hue='Rating color',data=ratings)
```

## mapping colours 

```python
sns.barplot(x="Aggregate rating",y='Rating count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:    

    1.Not rated count is very high    
    2.Maximum number of ratings are between 2.5-3.4    


```python
## Count plot
sns.countplot(x="Rating color",data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```
